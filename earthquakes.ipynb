```python
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
import matplotlib
matplotlib.style.use('ggplot')

df = pd.read_csv('query.csv')
df['time']=pd.to_datetime(df['time'])
df.set_index('time',inplace=True)
df = df[df.index>'2018-05']

fig = plt.figure(figsize=(15,15))
ax1 = fig.add_subplot(211)
ax1.plot(df.index,df.mag,marker='o',ms=4,ls="None",color='k')
ax1.set_xlabel('Date',fontsize=15)
ax1.set_ylabel('mag',fontsize=15)

ax2 = fig.add_subplot(212)
df.groupby(pd.TimeGrouper("H")).mean().mag.plot(ax=ax2, style='s',ls='-')
ax2.set_xlabel('Date',fontsize=15)
ax2.set_ylabel('Mean(mag)',fontsize=15)
plt.show()
```
